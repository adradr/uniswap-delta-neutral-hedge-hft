```python
import matplotlib.pyplot as plt
from uniswap_fee_and_divergence import *

uniswap_data = pd.read_csv("UNISWAP3POLYGON_USDCWETH_1H.csv")
uniswap_data.index = pd.to_datetime(uniswap_data.time, unit="s") # type: ignore
uniswap_data = uniswap_data[["open", "high", "low", "close", "Volume"]]
# Slice due to missing data 
uniswap_data = uniswap_data["2022-05-02 00:00:00":]
# Fill missing data
uniswap_data = uniswap_data.fillna(method="ffill")
# Resample to 1H and interpolate
uniswap_data = uniswap_data.loc[~uniswap_data.index.duplicated(keep='first')]
uniswap_data = uniswap_data.resample('H').interpolate(method='linear')
uniswap_data_indexes = uniswap_data.index
uniswap_data.head()
```

```python
# Create a unix timestamp from a string datetime
# from_date = "2022-06-08"
# to_date = "2023-06-08"

from_date = uniswap_data.index[0].strftime("%Y-%m-%d")
to_date = uniswap_data.index[-1].strftime("%Y-%m-%d")
print(from_date, to_date)


datamanger = DataFetcher(
    from_date=from_date,
    to_date=to_date,
    symbol="ETH/USDT",
    )

datamanger.fetch_data()

# Fill missing data
datamanger.df = datamanger.df.fillna(method="ffill")
# Resample to 1H and interpolate
datamanger.df = datamanger.df.loc[~datamanger.df.index.duplicated(keep='first')]
datamanger.df = datamanger.df.resample('H').interpolate(method='linear')
# Slice to match uniswap data
datamanger.df = datamanger.df.loc[uniswap_data_indexes[0]:uniswap_data_indexes[-1]]

# Replace datamaner.df.volume with uniswap_data.Volume
datamanger.df.volume = uniswap_data.Volume
```

```python
datamanger.df.to_csv("Backtest_ready_ETHUSDT_1H.csv", index=True)
```
